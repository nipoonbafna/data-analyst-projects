```python
# Customer Churn Analysis

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset
data = {
    "CustomerID": range(1, 21),
    "Tenure": [1, 5, 10, 15, 20, 2, 8, 12, 18, 6, 3, 14, 9, 7, 4, 16, 11, 13, 17, 19],
    "MonthlyCharges": [25, 50, 80, 100, 60, 30, 45, 70, 90, 55, 40, 85, 75, 35, 20, 95, 65, 78, 88, 99],
    "Churn": ["Yes", "No", "No", "Yes", "No", "Yes", "No", "No", "Yes", "No",
              "Yes", "No", "Yes", "No", "Yes", "No", "Yes", "No", "No", "Yes"]
}

df = pd.DataFrame(data)

# Churn rate
churn_rate = df["Churn"].value_counts(normalize=True) * 100
print("Churn Rate (%):\n", churn_rate)

# Visualization
sns.countplot(x="Churn", data=df)
plt.title("Customer Churn Distribution")
plt.show()

sns.boxplot(x="Churn", y="MonthlyCharges", data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```
