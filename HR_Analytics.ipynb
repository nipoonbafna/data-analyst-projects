```python
# HR Analytics Project

import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    "EmployeeID": range(1, 21),
    "Department": ["HR", "IT", "Finance", "Sales", "IT"] * 4,
    "Age": [22, 35, 40, 29, 31, 28, 32, 45, 41, 27, 30, 26, 34, 39, 33, 25, 37, 42, 36, 29],
    "Salary": [30000, 50000, 70000, 45000, 55000, 32000, 48000, 72000, 68000, 40000,
               35000, 42000, 53000, 61000, 47000, 31000, 59000, 64000, 56000, 46000],
    "Attrition": ["Yes", "No", "No", "Yes", "No", "Yes", "No", "No", "Yes", "No",
                  "No", "Yes", "No", "No", "Yes", "Yes", "No", "No", "Yes", "No"]
}

df = pd.DataFrame(data)

# Attrition by department
dept_attrition = df.groupby("Department")["Attrition"].value_counts().unstack().fillna(0)
print("Attrition by Department:\n", dept_attrition)

# Visualization
dept_attrition.plot(kind="bar", stacked=True, figsize=(8,5))
plt.title("Attrition by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.show()

# Export for Excel Dashboard
df.to_csv("hr_data.csv", index=False)
```
